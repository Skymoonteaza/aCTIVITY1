```python
# Calculate total sales for each product
df['Total Sales'] = df['Quantity'] * df['Price']

# Group by 'Product' and calculate the total sales per product
product_sales = df.groupby('Product')['Total Sales'].sum().reset_index()

# Sort the products by total sales in descending order
top_5_products = product_sales.sort_values(by='Total Sales', ascending=False).head(5)

top_5_products
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Grouping data by product to calculate total sales for each product
total_sales = df.groupby('Product')['Total'].sum().reset_index()

# Plotting the bar chart for total sales by product
plt.figure(figsize=(12, 6))
bar_colors = ['olive', 'salmon', 'lightblue', 'lightgreen', 'purple', 'pink', 'orange', 'lightcoral', 'lightyellow', 'red', 'green', 'blue', 'teal', 'brown', 'cyan', 'magenta', 'gray', 'indigo', 'yellowgreen', 'slateblue', 'darkblue', 'gold']
plt.bar(total_sales['Product'], total_sales['Total'], color=bar_colors)

# Adding titles and labels
plt.title('Total Sales for Each Product', fontsize=16)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert 'Date' column to datetime and extract month
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')

# Calculate the total sales for each product per month
monthly_sales = df.groupby(['Month', 'Product'])['Total'].sum().reset_index()

# Plotting the line chart for monthly sales trend by product
plt.figure(figsize=(12, 8))

# Defining colors for each product (you can modify the color list if needed)
color_map = {
    'Laptop': 'olive', 
    'Sofa': 'salmon',
    'T-Shirt': 'lightblue',
    'Smartphone': 'lightgreen',
    'Dining Table': 'purple',
    'Jacket': 'pink',
    'Headphones': 'orange',
    'Bed': 'lightcoral',
    'Shoes': 'lightyellow',
    'Monitor': 'red',
    'Dress': 'green',
    'Chair': 'blue',
    'Washing Machine': 'teal',
    'Bookshelf': 'brown',
    'Jeans': 'cyan',
    'Laptop Bag': 'magenta',
    'Tablet': 'gray',
    'Coffee Table': 'indigo',
    'Clothing': 'yellowgreen',
    'Furniture': 'slateblue',
    'Electronics': 'gold'
}

# Loop through each product to plot its sales trend
for product in monthly_sales['Product'].unique():
    product_sales = monthly_sales[monthly_sales['Product'] == product]
    plt.plot(product_sales['Month'].astype(str), product_sales['Total'], marker='o', color=color_map.get(product, 'black'), label=product)

# Adding titles, labels, and legend
plt.title('Monthly Sales Trend by Product', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title="Products", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Pie chart for Total Sales by Product
product_sales = df.groupby('Product')['Total'].sum().reset_index()

# Plot Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(product_sales['Total'], labels=product_sales['Product'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Total Sales by Product', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Key Insights 💖

From the pie chart above, we can see how the total sales are distributed across the different products. Here's a summary of the key insights:

- **Top Contributors**: The products contributing the highest to total sales are clearly visible, allowing us to identify which items performed the best in terms of sales.
- **Sales Distribution**: This pie chart visually represents how sales are split among the different product categories.
